```python
from elasticsearch import Elasticsearch
import json
```

```python
es = Elasticsearch("localhost:9200")
```

```python
with open('./Scraping virgool/VirgoolPosts.txt', 'r', encoding="utf8") as fi:
    data = json.load(fi)
```

```python
for i, d in enumerate(data):
    res = es.index(index='virgool-index', id=i, body=d)
```

```python
res
```

```python
try:
    res2 = es.get(index="virgool-index", id=9000000)
    print(res2)
except :
    print("not found")
```

```python
# , doc_type=None, params=None, headers=None
search_param = {
    'query': {
        'match': {
            'text': 'select device'
        }
    }
}
res = es.search(body=search_param, index="virgool-index")
```

```python
for i in res["hits"]["hits"]:
    print(i['_source']['title'])
```
